```python
import pandas as pd
```

```python
merged_data = pd.read_excel('2021jeju.xlsx')
merged_data
```

```python
range(len(merged_data))
for i in range(len(merged_data)):
    selected_data = merged_data.iloc[i]
    print(selected_data['VHCLE_NM'])
    print(selected_data['VHCLE_TY_NM'])
    print(selected_data['VHCLE_NMPR_CO'])
    print(selected_data['VHCLE_MAKR_NM'])
    print(selected_data['USE_TIME_CO'])
    print("-------------------------")

import pymysql
# MySQL 연결에 필요한 정보를 입력합니다.
con = pymysql.connect(host='localhost', user='root', password='mysql', db='pyprac2', charset='utf8',
                     cursorclass=pymysql.cursors.DictCursor) # 컬럼명까지 받아올지 여부
# SQL구문을 받아주는 cursor 객체를 사용합니다.
cur = con.cursor()
cur.execute("SELECT * FROM rent_tbl")

for i in range(len(merged_data)):
    selected_row = merged_data.iloc[i]
    cur.execute("INSERT INTO rent_tbl (VHCLE_NM, VHCLE_TY_NM, VHCLE_NMPR_CO, VHCLE_MAKR_NM, USE_TIME_CO) VALUES ('%s', '%s', '%s', '%s', %s)" % (selected_row['VHCLE_NM'], selected_row['VHCLE_TY_NM'], selected_row['VHCLE_NMPR_CO'], selected_row['VHCLE_MAKR_NM'], selected_row['USE_TIME_CO']))
con.commit()
```

```python
cur.execute("SELECT * FROM rent_tbl")
mysql_data = cur.fetchall()
mysql_data 
newData = pd.DataFrame(mysql_data)
newData
```

```python
group = newData.groupby('VHCLE_NM')
newGroup = group.size()
newGroups = pd.DataFrame(newGroup)
newGroups
```

```python
newGroupz = newGroups.rename(columns={0:'USE_TIME_CO'})
exportData = newGroupz.sort_values(by='USE_TIME_CO', ascending=False)
exportData
```

```python
exportData.to_csv('rentCar.csv')
```
